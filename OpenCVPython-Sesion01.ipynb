##FORMACIÃ“N DE IMAGENES

Mostrar una imagen

```
import cv2
import numpy as np
from skimage import io
from google.colab.patches import cv2_imshow

url = "https://e00-marca.uecdn.es/assets/multimedia/imagenes/2020/09/21/16006800371103.jpg"
image = io.imread(url)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2_imshow(image)
```

Acceder a los valores de un pixel

```
rows, cols, channels = image.shape
print(f"Filas: {rows}")
```

```
for i in range (rows):
  for j in range (1,2):
    value = image[i, j]
    print(value)
```

Extraer canales de una imagen

```
url2 = "https://pbs.twimg.com/profile_images/690552069145538560/90NPsfbv.jpg"
image2 = io.imread(url2)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

cv2_imshow(image2)
```

```
b = image2[:,:,0]
g = image2[:,:,1]
r = image2[:,:,2]
cv2_imshow(r)
concat_image = cv2.hconcat((r,b))
print("")
cv2_imshow(concat_image)
```

Convertir imagen RGB a Gray u otro canal de color

```
url3 = "https://guias-viajar.com/wp-content/uploads/2019/05/groenlandia-auroras-boreales-javier-alippi.jpg"
image3 = io.imread(url3)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

cv2_imshow(image3)

gray = cv2.cvtColor(image3, cv2.COLOR_RGB2GRAY)

cv2_imshow(gray)
```

```
color_space = cv2.cvtColor(image3, cv2.COLOR_RGB2HSV)
cv2_imshow(color_space)
```

##PROCESAMIENTO DE IMAGENES

Filtros lineales: Filtro de media

```
url4 = "https://www.fit.vutbr.cz/~vasicek/imagedb/img_corrupted/impnoise_005/106020.png"
image4 = io.imread(url4)
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)
image4 = cv2.cvtColor(image4, cv2.COLOR_RGB2GRAY)

cv2_imshow(image4)
```

```
processed_image = cv2.medianBlur(image4, 3)
cv2_imshow(processed_image)
```

Filtros lineales: Filtro gaussiano

```
url5 = "https://people.sc.fsu.edu/~jburkardt/c_src/image_denoise/balloons_noisy.png"
image5 = io.imread(url5)
image5 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)

cv2_imshow(image5)
```

```
procesed_image = cv2.GaussianBlur(image5,(5,5),0)
cv2_imshow(procesed_image)
```
